# <h1> TASK-1 To Predict the percentage of marks of the students based on the number of hours they studied.<h1>

```python

# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv('http://bit.ly/w-data')
data.head(5)
```

```python
data.isnull == True
```

<b> There is no null value in the Dataset so, we can now visualize our Data.<b>

```python
sns.set_style('darkgrid')
sns.scatterplot(y= data['Scores'], x= data['Hours'])
plt.title('Marks Vs Study Hours',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

<b> From the above scatter plot there looks to be correlation between the 'Marks Percentage' and 'Hours Studied', Lets plot a regression line to confirm the correlation.</b>

```python
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
print(data.corr())
```

<b><i> It is confirmed that the variables are positively correlated.</i></b>

<h1><b1> Training the Model </h1></b1>

# <h1>1) Splitting the Data</h1>

```python
# Defining X and y from the Data
X = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values

# Spliting the Data in two
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```


<h1>2) Fitting the Data into the model</h1>

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("---------Model Trained---------")
```

<h1><i><b>Predicting the Percentage of Marks</b></i><h1>

```python
pred_y = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

# <h1><i><b>Comparing the Predicted Marks with the Actual Marks</b></i><h1>

```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

# <h1><i><b>Visually Comparing the Predicted Marks with the Actual Marks</b></i><h1>

```python
plt.scatter(x=val_X, y=val_y, color='blue')
plt.plot(val_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

# <h1><i><b>Evaluating the Model</b></i><h1>

```python
# Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

# <h1>Small value of Mean absolute error states that the chances of error or wrong forecasting through the model are very less.</h1>

# <h1><i><b>What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?</b></i></h1>

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

# According to the regression model if a student studies for 9.25 hours a day he/she is likely to score 93.89 marks.
